```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import os
```

```python
#cwd = os.getcwd()
#print(cwd)

# Variables
file = '../data/H-H2_LOSC_4_V1-875225088-4096.hdf5'
data = h5py.File(file, 'r')
group = data['strain']

for key in group.keys():
    print("The key(s) for group is --> " + key)
    
```

```python
num_samples = 1000000
```

```python
strain = data['strain']['Strain'].value
time = np.arange(0, 1, 1/num_samples)

plt.plot(time, strain[:num_samples])
plt.xlabel('GPS Time (s)')
plt.ylabel('key: strain')
plt.show()
```
